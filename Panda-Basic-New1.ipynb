```python
df2 = df.rename(columns={'Acceleration (0-60) Sec':'Acceleration'})
df2
```

```python
#Column Subsetting (also knownas as column slicing)

df2[['Overall Score']] #double brackets returns an object of type dataframe
#data frame with one single column

#On the left side --> 0,1,2,3,4.... --> index --> is not a number
#Index --> a label (string)

mean_overall_score = df2[['Overall Score']].mean()
display(mean_overall_score)

median_overall_score = df2[['Overall Score']].median()
display(median_overall_score)

mode_overall_score = df2[['Overall Score']].mode()
display(mode_overall_score)
```
